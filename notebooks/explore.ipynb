```python
#example folium chloropleth map
import folium
map = folium.Map(location=[59.31544, 18.069029], tiles='cartodbpositron', zoom_start=11)

fig = folium.Choropleth(geo_data=gj_neighbourhoods, key_on="properties[neighbourhood]").add_to(map)

map
```

```python
import plotly.express as px

fig = px.choropleth_mapbox(
                        data_frame=df_neighbourhood_overview,
                        geojson=gj_neighbourhoods,
                        featureidkey="properties.neighbourhood",
                        locations="neighbourhood",
                        color="availability_365",
                        color_continuous_scale="viridis",
                        mapbox_style="carto-positron",
                        zoom=10,
                        center = {"lat": 59.31544, "lon": 18.069029}, 
                        opacity=0.5,
                        )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
